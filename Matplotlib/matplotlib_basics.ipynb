```python
# Bar Graph
# When data = categories + numbers
# Examples:

# students vs marks

# product vs sales

subjects = ["Math","Physics","Chemistry", "English"]
marks = [89, 90,76, 85]
plt.bar(subjects, marks, color="red", edgecolor="yellow")

plt.bar(subjects, marks)
plt.title("marks in subjects")
plt.xlabel("subjects")
plt.ylabel("marks")
plt.show()
```

```python
#SCATTER PLOT  Used to check correlation.
# If dots go upward → positive correlation
x = [1,2,3,4,5,6,7]
y = [2,4,5,6,7,2,3]

plt.scatter(x, y)
plt.xlabel("Hours Studied")
plt.ylabel("marks")
plt.show()
```

```python
# PIE CHART — Percentage Contribution

labels = ["Study", "Sleep", "Mobile", "Gym"]
hours = [6,8,7,3]

plt.pie(hours, labels=labels, autopct="%1.1f%%")
plt.show()
```

```python

# ----------------------------------------------------practice----------------------------------------------------------

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv(r"C:\Users\Hp\OneDrive\Desktop\AIML\studentdata.csv")
plt.figure(figsize=(14,4))  


# plt.subplot(rows, columns, position)
plt.subplot(1,3,1)
plt.hist(df["math score"], bins=15, edgecolor="black")
plt.title("math score distribution")
plt.xlabel("Marks")
plt.ylabel("Students")
plt.grid(alpha=0.3)

plt.subplot(1,3,2)
plt.hist(df["reading score"], bins = 15, edgecolor="black")
plt.title("reading Score Distribution")
plt.xlabel("Marks")
plt.ylabel("Students")
plt.grid(alpha=0.3)

plt.subplot(1,3,3)
df["gender"].value_counts().plot(kind="bar")
plt.title("Gender count")
plt.xlabel("Gender")
plt.ylabel("Students")
plt.tight_layout()  #It adjusts: margins, padding, spacing between plots

plt.savefig("report.jpg", dpi=300, bbox_inches="tight")   #DPi:- Dot per Inch(image quality) bbox_inches = Crops unnecessary blank space
plt.show()




```
