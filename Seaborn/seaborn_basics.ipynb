```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv(r"C:\Users\Hp\OneDrive\Desktop\AIML\studentdata.csv")

sns.set_theme()
sns.set_style("white")
# in seaborn we only mention the column name and it will take auto like count and the value 
sns.histplot(data = df, x = "math score", bins = 20, color="purple")
# plt.savefig("demo.jpg", dpi=300, bbox_inches="tight")
plt.show()


# ----------------------------Multiple Column Direct Plot (Powerful)---------------------------------------


sns.set_theme()
# sns.set_style("darkgrid")
plt.title("Math Score Distribution by Gender")
sns.histplot(data = df, x = "math score", hue="gender", bins =20)  #hue (group the data inside the same graph based on a category)
plt.show() 

```

```python
# PHASE 2 Distribution Analysis 

# histplot() — Improved Histogram (recap + upgrade)
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv(r"C:\Users\Hp\OneDrive\Desktop\AIML\studentdata.csv")

sns.set_theme()
sns.set_style("darkgrid")
sns.histplot(data = df, x = "math score", bins = 20, kde = True) #Kernel Density Estimation
plt.show()

# --------------------------------kdeplot() — Smooth Distribution Curve----------------------
sns.kdeplot(data = df, x = "math score", hue="gender", fill=True)  #A smooth probability curve instead of bars.
plt.show() 

# -------------------------------boxplot() — Outlier Detector-------------------------------
sns.boxplot(data = df, x="math score")  #Dots outside = outliers
plt.show()

# -------------------violinplot() — Distribution + Density Together-(Shows: median, spread, density shape)------------------------

sns.violinplot(data = df, x="gender",y="math score")
plt.show()
```

```python
# PHASE 3 - Feature Relationships (Correlation Understanding)
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv(r"C:\Users\Hp\OneDrive\Desktop\AIML\studentdata.csv")
#------------------------------scatterplot() — Basic Relationship---------------------------
sns.scatterplot(data=df, x="math score", y="reading score", hue ="gender")
plt.title("math score and reading score distribution")
plt.show()

#-----------------------------------regplot()- utomatically draws regression line.----------------------------------------------

sns.regplot(data = df, x="math score", y = "reading score")
# # NOTE:- regplot doesn't directly support hue we use implot 
sns.lmplot(data = df, x="math score", y = "reading score", hue="gender")
plt.show()


# ---------------------------------pairplot() — Entire Dataset Relationships --------------------------------------

sns.pairplot(df)
plt.show()

sns.pairplot(df, hue="gender")
plt.show()
```
