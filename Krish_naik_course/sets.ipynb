```python
# sets are the built-in data type in python used to store collections of unique data and they do not allow duplicate elements, sets are useful for membership tests, eliminating duplicate entries and performing mathematical set operations like union, intersection, difference and symetric difference 

```

```python
##create a set 
my_set = {1,2,3,4,5}
print(my_set)
print(type(my_set))

my_empty_set = set([1,2,3,4,5])
print(my_empty_set)
```

```python
##basic set operations 

#adding and removing elements

my_set.add(20)
print(my_set)

# my_set.remove(20)
my_set.discard(39)
print(my_set)



```

```python
##pop elements
removed_element = my_set.pop()
print(my_set)

```

```python
##clear all the elemnents

my_set.clear()
print(my_set)
```

```python
##set membership test
my_set = {1,2,3,4,5}
print(3 in my_set)
print(14 in my_set) #14 is not present it will give false
```

```python
##mathematical operations

set1 = {1,2,3,4,5}
set2={6,7,8,9,10}
union_set = set1.union(set2)
print(union_set)

#intersection

intersection_set = set1.intersection(set2)
print(intersection_set)


set1.intersection_update(set2)
print(set1)
```
